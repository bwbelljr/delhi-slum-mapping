```python
# Place here so that I can move this Jupyter file to wherever I need

def generate_tif_file_list(rootdir):
    # Description: Given root directory, returns list
    # of all .tif files
    # Adapted from https://stackoverflow.com/questions/31948528/python-loop-through-files-of-certain-extensions

    import os
    extensions = ('.tif')
    tif_file_list = []

    for subdir, dirs, files in os.walk(rootdir):
        for file in files:
            ext = os.path.splitext(file)[-1].lower()
            if ext in extensions:
                file_location = os.path.join(subdir, file)
                if '.tif' in file_location.lower(): # ensures ending is .tif
                    tif_file_list.append(file_location)

    return (tif_file_list)
```

```python
def generate_RGB_from_dg(rootdir):
    # Description: given directory, iterate through .tif files
    # and convert each to RGB image by extracting only the first
    # 3 bands and stacking them into a new GeoTiff file.
    
    # Assumes generate_tif_file_list is defined or imported
    file_list = generate_tif_file_list(rootdir)
    
    for tif_file in file_list:
        
        # Find where actual filename begins
        actual_tif_file = tif_file[len(rootdir)+1:]
        new_tif_file = actual_tif_file[:-4]+'_rgb.tif'
        print(actual_tif_file)
        print(new_tif_file)
        print('\n')
        # run command line in script, but using Python filenames
        ! rio stack $actual_tif_file --bidx 1,2,3 $new_tif_file --profile interleave=band
    
    print('completed all files')
```
